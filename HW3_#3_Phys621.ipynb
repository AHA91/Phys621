```python
###3a

import numpy as np
import matplotlib.pyplot as plt

L = np.arange(10**-6,0.001,10**-6)

ni = 0.75
ai = 800
R1 = 0.3
R2 = 0.3
R = R1*R2

n = ni*((1/(2*L))*np.log(1/R))/(ai + (1/(2*L))*np.log(1/R))

plt.plot(L,1/n)
plt.xlabel("Length (m)")
plt.ylabel("External differential quantum efficiency")
plt.title(" 3a - Inverse External Differentail Quantum Efficiency Vs Length")
plt.plot()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

If = np.linspace(0.003,0.01,1000)

Lb = 0.0005
Lc = 0.001
ni = 0.75
ai = 800
R1 = 0.3
R2 = 0.3
R = R1*R2
Ithb = 0.0035
Ithc = 0.004

################3b###############
nb = ni*((1/(2*Lb))*np.log(1/R))/(ai + (1/(2*Lb))*np.log(1/R))
ns = 0.62*nb/1.55
Pout_b = ns*(If-Ithb)

################3c###############
nc = ni*((1/(2*Lc))*np.log(1/R))/(ai + (1/(2*Lc))*np.log(1/R))
ns = 0.62*nc/1.55
Pout_c = ns*(If-Ithc)


plt.plot(If,Pout_b, label = "3b - L=500um, Ith = 3.5mA")
plt.plot(If, Pout_c, label = "3c - L=1000um, Ith = 4mA")
plt.xlabel("If (A)")
plt.ylabel("Power output (Watts)")
plt.title("Output Power vs Threshold Current")
plt.ylim([0,0.00150])
plt.legend()
plt.show()
```
